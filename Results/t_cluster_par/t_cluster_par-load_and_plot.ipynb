```python
import numpy
from numpy import array, mean, std, concatenate
import pickle
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='serif'
matplotlib.rcParams['axes.unicode_minus']=False
matplotlib.rcParams.update({'font.size': 20})
matplotlib.rc('text', usetex=True)
```

```python
path = "/home/investigator/Work/qcnn-var/Results/t_cluster_par/"
```

```python
# info = "t_cluster_par" + "-qcnn_cluster_9q-" + ("n=%d-c=%d-m=%s-r=%d-w_ls=%.1f-w_var=%.4f" %(9, 1, [4], 10, 1, 1e-4)).replace(" ", "")
info = "t_cluster_par" + "-qcnn_cluster_9q-" + ("n=%d-c=%d-m=%s-r=%d-w_ls=%.1f-w_var=%.4f" %(9, 1, [4], 30, 1, 1e-4)).replace(" ", "")
print(path + info)
with open(path + info + ".dict", 'rb') as fp:
    data_dict_loaded = pickle.load(fp)

QFIs = data_dict_loaded["QFIs"]
labels_test = data_dict_loaded["labels_test"]

disps_mean_qcnn_4 = mean(data_dict_loaded["disps_list"], axis=0)
disps_red_mean_qcnn_4 = mean(data_dict_loaded["disps_list"]/data_dict_loaded["expecs_der_list"]**2, axis=0)
disps_red_std_qcnn_4 = std(data_dict_loaded["disps_list"]/data_dict_loaded["expecs_der_list"]**2, axis=0)
biass_sq_mean_qcnn_4 = mean((data_dict_loaded["expecs_list"] - data_dict_loaded["labels_test"])**2, axis=0)
biass_sq_std_qcnn_4 = std((data_dict_loaded["expecs_list"] - data_dict_loaded["labels_test"])**2, axis=0)
CFIs_mean_qcnn_4 = mean(data_dict_loaded["CFIs_list"], axis=0)

biass_sq_fill_l_qcnn_4 = array(biass_sq_mean_qcnn_4 - biass_sq_std_qcnn_4)
biass_sq_fill_l_qcnn_4 = array([biass_sq_fill_l_qcnn_4[i] if biass_sq_fill_l_qcnn_4[i] > 0 else 0 for i in range(len(biass_sq_fill_l_qcnn_4))])
biass_sq_fill_u_qcnn_4 = array(biass_sq_mean_qcnn_4 + biass_sq_std_qcnn_4)
disps_red_fill_l_qcnn_4 = array(disps_red_mean_qcnn_4 - disps_red_std_qcnn_4)
disps_red_fill_l_qcnn_4 = array([disps_red_fill_l_qcnn_4[i] if disps_red_fill_l_qcnn_4[i] > 1/QFIs[i] else 1/QFIs[i] for i in range(len(disps_red_fill_l_qcnn_4))])
disps_red_fill_u_qcnn_4 = array(disps_red_mean_qcnn_4 + disps_red_std_qcnn_4)
```

```python
# info = "t_cluster_par" + "-qcnn_cluster_9q-" + ("n=%d-c=%d-m=%s-r=%d-w_ls=%.1f-w_var=%.4f" %(9, 1, [3,4,5], 10, 1, 1e-4)).replace(" ", "")
info = "t_cluster_par" + "-qcnn_cluster_9q-" + ("n=%d-c=%d-m=%s-r=%d-w_ls=%.1f-w_var=%.4f" %(9, 1, [1, 4, 7], 30, 1, 1e-4)).replace(" ", "")
print(path + info)
with open(path + info + ".dict", 'rb') as fp:
    data_dict_loaded = pickle.load(fp)

disps_mean_qcnn147 = mean(data_dict_loaded["disps_list"], axis=0)
disps_red_mean_qcnn147 = mean(data_dict_loaded["disps_list"]/data_dict_loaded["expecs_der_list"]**2, axis=0)
disps_red_std_qcnn147 = std(data_dict_loaded["disps_list"]/data_dict_loaded["expecs_der_list"]**2, axis=0)
biass_sq_mean_qcnn147 = mean((data_dict_loaded["expecs_list"] - data_dict_loaded["labels_test"])**2, axis=0)
biass_sq_std_qcnn147 = std((data_dict_loaded["expecs_list"] - data_dict_loaded["labels_test"])**2, axis=0)
CFIs_mean_qcnn147 = mean(data_dict_loaded["CFIs_list"], axis=0)

biass_sq_fill_l_qcnn147 = array(biass_sq_mean_qcnn147 - biass_sq_std_qcnn147)
biass_sq_fill_l_qcnn147 = array([biass_sq_fill_l_qcnn147[i] if biass_sq_fill_l_qcnn147[i] > 0 else 0 for i in range(len(biass_sq_fill_l_qcnn147))])
biass_sq_fill_u_qcnn147 = array(biass_sq_mean_qcnn147 + biass_sq_std_qcnn147)
disps_red_fill_l_qcnn147 = array(disps_red_mean_qcnn147 - disps_red_std_qcnn147)
disps_red_fill_l_qcnn147 = array([disps_red_fill_l_qcnn147[i] if disps_red_fill_l_qcnn147[i] > 1/QFIs[i] else 1/QFIs[i] for i in range(len(disps_red_fill_l_qcnn147))])
disps_red_fill_u_qcnn147 = array(disps_red_mean_qcnn147 + disps_red_std_qcnn147)
```

```python
# info = "t_cluster_par" + "-hva_transverse_cluster-l=7-" + ("n=%d-c=%d-m=%s-r=%d-w_ls=%.1f-w_var=%.4f" %(9, 1, [3,4,5], 10, 1, 1e-4)).replace(" ", "")
info = "t_cluster_par" + "-hva_transverse_cluster-" + ("n=%d-c=%d-m=%s-r=%d-w_ls=%.1f-w_var=%.4f-l=%d" %(9, 1, [1,4,7], 30, 1, 1e-4, 10)).replace(" ", "")
print(path + info)
with open(path + info + ".dict", 'rb') as fp:
    data_dict_loaded = pickle.load(fp)

disps_mean_hva147 = mean(data_dict_loaded["disps_list"], axis=0)
disps_red_mean_hva147 = mean(data_dict_loaded["disps_list"]/data_dict_loaded["expecs_der_list"]**2, axis=0)
disps_red_std_hva147 = std(data_dict_loaded["disps_list"]/data_dict_loaded["expecs_der_list"]**2, axis=0)
biass_sq_mean_hva147 = mean((data_dict_loaded["expecs_list"] - data_dict_loaded["labels_test"])**2, axis=0)
biass_sq_std_hva147 = std((data_dict_loaded["expecs_list"] - data_dict_loaded["labels_test"])**2, axis=0)
CFIs_mean_hva147 = mean(data_dict_loaded["CFIs_list"], axis=0)

biass_sq_fill_l_hva147 = array(biass_sq_mean_hva147 - biass_sq_std_hva147)
biass_sq_fill_l_hva147 = array([biass_sq_fill_l_hva147[i] if biass_sq_fill_l_hva147[i] > 0 else 0 for i in range(len(biass_sq_fill_l_hva147))])
biass_sq_fill_u_hva147 = array(biass_sq_mean_hva147 + biass_sq_std_hva147)
disps_red_fill_l_hva147 = array(disps_red_mean_hva147 - disps_red_std_hva147)
disps_red_fill_l_hva147 = array([disps_red_fill_l_hva147[i] if disps_red_fill_l_hva147[i] > 1/QFIs[i] else 1/QFIs[i] for i in range(len(disps_red_fill_l_hva147))])
disps_red_fill_u_hva147 = array(disps_red_mean_hva147 + disps_red_std_hva147)
```

```python
# matplotlib.rcParams['mathtext.fontset'] = 'stix'
# matplotlib.rcParams['font.family'] = 'STIXGeneral'

info = "t_cluster_par-" + ("n=%d-c=%d-r=%d-w_ls=%.1f-w_var=%.4f" %(9, 1, 10, 1, 1e-4)).replace(" ", "")

plt.figure(figsize=(6.25, 4.25))
# plt.scatter(labels_train, biass_train**2, s=100)
p_qcnn_4 = plt.plot(labels_test, biass_sq_mean_qcnn_4, color="tab:blue", linewidth=2)
p_qcnn147 = plt.plot(labels_test, biass_sq_mean_qcnn147, color="tab:red", linewidth=2)
p_hva147 = plt.plot(labels_test, biass_sq_mean_hva147, color="gold", linewidth=2)
f_qcnn_4 = plt.fill_between(labels_test, biass_sq_fill_l_qcnn_4, biass_sq_fill_u_qcnn_4, alpha=0.3, color="tab:blue")
f_qcnn147 = plt.fill_between(labels_test, biass_sq_fill_l_qcnn147, biass_sq_fill_u_qcnn147, alpha=0.3, color="tab:red")
f_hva147 = plt.fill_between(labels_test, biass_sq_fill_l_hva147, biass_sq_fill_u_hva147, alpha=0.3, color="gold")
plt.xlabel(r"$x$")
plt.ylabel(r"$(\mathsf{x} - x)^2$")
# plt.legend()
plt.legend([(p_qcnn_4[0], f_qcnn_4), (p_qcnn147[0], f_qcnn147), (p_hva147[0], f_hva147)],
           [r'QCNN, $m=1$', r'QCNN, $m=3$', r'HVA, $m=3$'], framealpha=1, loc=2) #, bbox_to_anchor=(0.475, 1)
# plt.yscale("log")
plt.grid()
plt.tight_layout()
# plt.savefig(info + "-est.pdf", transparent=True, bbox_inches='tight')
plt.show()


sss = 145
fff = 157
mmm = 150
li_4 = concatenate([disps_red_mean_qcnn_4[:sss], [disps_red_mean_qcnn_4[mmm]], disps_red_mean_qcnn_4[fff:]])
li147 = concatenate([disps_red_mean_qcnn147[:sss], [disps_red_mean_qcnn147[mmm]], disps_red_mean_qcnn147[fff:]])
fb_l_4 = concatenate([disps_red_fill_l_qcnn_4[:sss], [disps_red_fill_l_qcnn_4[mmm]], disps_red_fill_l_qcnn_4[fff:]])
fb_u_4 = concatenate([disps_red_fill_u_qcnn_4[:sss], [disps_red_fill_u_qcnn_4[mmm]], disps_red_fill_u_qcnn_4[fff:]])
fb_l147 = concatenate([disps_red_fill_l_qcnn147[:sss], [disps_red_fill_l_qcnn147[mmm]], disps_red_fill_l_qcnn147[fff:]])
fb_u147 = concatenate([disps_red_fill_u_qcnn147[:sss], [disps_red_fill_u_qcnn147[mmm]], disps_red_fill_u_qcnn147[fff:]])
lt = concatenate([labels_test[:sss], [labels_test[mmm]], labels_test[fff:]])
plt.figure(figsize=(6.25, 4.25))
plt.plot(lt, li_4, color="tab:blue", linewidth=2)
plt.plot(lt, li147, color="tab:red", linewidth=2)
plt.plot(labels_test, disps_red_mean_hva147, color="gold", linewidth=2)
plt.fill_between(lt, fb_l_4, fb_u_4, alpha=0.3, color="tab:blue")
plt.fill_between(lt, fb_l147, fb_u147, alpha=0.3, color="tab:red")
plt.fill_between(labels_test, disps_red_fill_l_hva147, disps_red_fill_u_hva147, alpha=0.3, color="gold")
plt.plot(labels_test, 1/QFIs, color="tab:green", label=r"$\frac{1}{I_q(\psi_x)}$", linewidth=5, linestyle="solid") # , color="tab:blue"
plt.ylabel(r"$\frac{\Delta^2 M^*_m}{|\partial_x \langle M^*_m \rangle |^2}$" )
plt.xlabel(r"$x$")
plt.grid()
plt.yscale("log")
plt.legend(framealpha=1)
plt.tight_layout()
# plt.savefig(info + "-var.pdf", transparent=True, bbox_inches='tight')
plt.show()
```
