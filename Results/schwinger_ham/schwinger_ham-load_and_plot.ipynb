```python
import numpy
from numpy import array, mean, std
import pickle
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='serif'
matplotlib.rcParams['axes.unicode_minus']=False
matplotlib.rcParams.update({'font.size': 20})
matplotlib.rc('text', usetex=True)
```

```python
n_inp = 8
n_copies = 1
n_runs = 10
w_ls = 1
w_var = 1e-4
ham_name = "schwinger_ham"

path = "/home/investigator/Work/qcnn-var/Results/schwinger_ham/"
```

```python
n_meas_list = [0]
ansatz_name = "qcnn_schwinger"
folder = "/"

info = ham_name + "-" + ansatz_name + ("-n=%d-c=%d-m=%s-r=%d-w_ls=%.1f-w_var=%.4f" %(n_inp, n_copies, n_meas_list, n_runs, w_ls, w_var)).replace(" ", "")

print(path + folder + info)
with open(path + folder + info + ".dict", 'rb') as fp:
    data_dict_loaded = pickle.load(fp)

QFIs = data_dict_loaded["QFIs"]
labels_test = data_dict_loaded["labels_test"]

disps_mean_qcnn_0 = mean(data_dict_loaded["disps_list"], axis=0)
disps_red_mean_qcnn_0 = mean(data_dict_loaded["disps_list"]/data_dict_loaded["expecs_der_list"]**2, axis=0)
disps_red_std_qcnn_0 = std(data_dict_loaded["disps_list"]/data_dict_loaded["expecs_der_list"]**2, axis=0)
biass_sq_mean_qcnn_0 = mean((data_dict_loaded["expecs_list"] - data_dict_loaded["labels_test"])**2, axis=0)
biass_sq_std_qcnn_0 = std((data_dict_loaded["expecs_list"] - data_dict_loaded["labels_test"])**2, axis=0)
CFIs_mean_qcnn_0 = mean(data_dict_loaded["CFIs_list"], axis=0)

biass_sq_fill_l_qcnn_0 = array(biass_sq_mean_qcnn_0 - biass_sq_std_qcnn_0)
biass_sq_fill_l_qcnn_0 = array([biass_sq_fill_l_qcnn_0[i] if biass_sq_fill_l_qcnn_0[i] > 0 else 0 for i in range(len(biass_sq_fill_l_qcnn_0))])
biass_sq_fill_u_qcnn_0 = array(biass_sq_mean_qcnn_0 + biass_sq_std_qcnn_0)
disps_red_fill_l_qcnn_0 = array(disps_red_mean_qcnn_0 - disps_red_std_qcnn_0)
disps_red_fill_l_qcnn_0 = array([disps_red_fill_l_qcnn_0[i] if disps_red_fill_l_qcnn_0[i] > 1/QFIs[i] else 1/QFIs[i] for i in range(len(disps_red_fill_l_qcnn_0))])
disps_red_fill_u_qcnn_0 = array(disps_red_mean_qcnn_0 + disps_red_std_qcnn_0)
```
